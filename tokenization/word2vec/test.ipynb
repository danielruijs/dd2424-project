```python
import os
import sys
sys.path.append("~/dd2424-project")
from tokenization.main import load_text
# Suppress TensorFlow Info messages
os.environ["TF_CPP_MIN_LOG_LEVEL"] = (
    "3"  # 0: all messages, 1: filter out INFO messages, 2: filter out WARNING messages, 3: filter out ERROR messages
)
from tensorflow.keras.preprocessing.text import Tokenizer
from nltk.tokenize import RegexpTokenizer
```

```python
text, _, _ = load_text("shubhammaindola/harry-potter-books")
```

```python
def tokenize_text(text):
    """
    Tokenizes the input text into words.
    """
    tokenizer = RegexpTokenizer(r"\b\w+'\w+|\w+|\S")
    tokens = tokenizer.tokenize(text)
    print("Number of tokens:", len(tokens))

    tokenizer = Tokenizer(lower=False, filters="")
    tokenizer.fit_on_texts(tokens)
    words2idx = tokenizer.word_index
    idx2words = tokenizer.index_word
    vocab_size = len(words2idx) + 1
    print("Vocabulary size:", vocab_size)
    print("idx2words:", list(idx2words.values())[:10000])
    print("words2idx:", list(words2idx.items())[:10000])

    return words2idx, idx2words, vocab_size

words2idx, idx2words, vocab_size = tokenize_text(text)
```

```python
print("Words containing ':", [word for word in words2idx if any(char == "'" for char in word)])
print("Number of words containing ':", len([word for word in words2idx if any(char == "'" for char in word)]))
print("Words containing special characters:", [word for word in words2idx if any(char in "!@#$%^&*()_+-=[]{}|;':\",.<>?/" for char in word)])
print("Number of words containing special characters:", len([word for word in words2idx if any(char in "!@#$%^&*()_+-=[]{}|;':\",.<>?/" for char in word)]))
print("Words ending with .:", [word for word in words2idx if word.endswith(".")])
print("Number of words ending with .:", len([word for word in words2idx if word.endswith(".")]))
```
